# Implementation of Linear Regression Model with one feature

## Problem Statement ##
Suppose you are the CEO of a restaurant franchise and are considering different cities for opening a new outlet.

* You would like to expand your business to cities that may give your restaurant higher profits.
* The chain already has restaurants in various cities and you have data for profits and populations from the cities.
* You also have data on cities that are candidates for a new restaurant.
For these cities, you have the city population.

*Can you use the data to help you identify which cities may potentially give your business higher profits?*

## Importing Necessary Libraries
* NumPy, a popular library for scientific computing
* Matplotlib, a popular library for plotting data

```python
import math
import numpy as np
import matplotlib.pyplot as plt
```

## Compute Cost

* This function is used to calculate the cost during the learning process.

```python
def compute_cost (x, y, w, b):
    """
    x = input/feature, ndarray, (m, )
    y = output/target, ndarray, (m, )
    w, b = scalar parameters
    """

    m = x.shape[0]  # no of training examples

    total_cost = 0.0

    for i in range (m):
        f_wb = w * x[i] + b
        err = f_wb - y[i]
        total_cost = total_cost + (err ** 2)

    total_cost = total_cost / (2 * m)

    return total_cost
```

## Compute Gradient
* This function is used to determine the derivatives dj_dw, dj_db

```python
def compute_gradient(x, y, w, b):
    """
    x, y = ndarray, (m, )
    w, b = scalar parameters
    """ 
    m = x.shape[0]     # no of training examples
    dj_dw = 0.0
    dj_db = 0.0

    for i in range (m):
        f_wb = w * x[i] + b
        err = f_wb - y[i]

        dj_db = dj_db + err
        dj_dw = dj_dw + err * x[i]

    dj_dw /= m
    dj_db /= m

    return dj_dw, dj_db
```

## Gradient Descent

```python
def gradient_descent(x, y, w_in, b_in, alpha, no_iterations):
    """
    x, y = ndarray, (m, )
    w_in, b_in = initial value of the parameters
    alpha = learning rate
    no_iterations = no of iterations we will run to find minimum cost, J
    """
    w = w_in
    b = b_in

    J_history = []  # a list to store cost for later plotting purpose
    w_history = []  # to save w values , used later in plot
    
    for i in range(no_iterations):
        # compute gradient 
        dj_dw, dj_db = compute_gradient(x, y, w, b)
        # update parameters w and b simultaneously
        w = w - alpha * dj_dw
        b = b - alpha * dj_db

        # save cost at each iterations
        if (i < no_iterations):
            cost = compute_cost (x, y, w, b)
            J_history.append(cost)
            
        # Print cost every at intervals 10 times or as many iterations if < 10
        if i% math.ceil(no_iterations/10) == 0:
            w_history.append(w)
            print(f"Iteration {i:4}: Cost {float(J_history[-1]):8.2f}   ")
        
    return w, b, J_history, w_history
```

## Step 1: Load data from text file

```python
data = np.loadtxt("data.txt", delimiter=',')

x_train = data[:,0]
y_train = data[:,1]
```

```python
# print x_train
print("Type of x_train:",type(x_train))
print("First five elements of x_train are:\n", x_train[:5])
```

```python
# print y_train
print("Type of y_train:", type(x_train))
print("First five elements of y_train are:\n", y_train[:5])
```

## Check the dimension of variable

```python
print("The shape of x_train is: ", x_train.shape)
print("The shape of y_train is: ", y_train.shape)
print("Number of training examples (m): ", len(x_train))
```

## Step 2: Visualize the dataset #

```python
# marker = 'x' means each point of the dataset represented as x
# c = 'r' means color of the point is red
plt.scatter(x_train, y_train, marker='x', c='r') 

# Set the title
plt.title("Profits vs. Population per city")
# Set the y-axis label
plt.ylabel('Profit in $10,000')
# Set the x-axis label
plt.xlabel('Population of City in 10,000s')
# show the plot
plt.show()
```

## Step 3: Initialize parameters ##

```python
w_init = 0.0
b_init = 0.0

no_of_iterations = 10000
alpha = 0.003
```

## Step 4: Run Gradient Descent Until Converge ##

```python
w_final, b_final, J_history, w_history = gradient_descent(x_train, y_train, w_init, b_init, alpha, no_of_iterations)

print("Final value of w and b are: ", w_final, b_final)
```

## Step 5: Cost VS Iterations Plot
*This plot help us to understand gradient descent properly working or not* 

```python
# plot cost versus iteration  
fig, (ax1, ax2) = plt.subplots(1, 2, constrained_layout=True, figsize=(12,4))
ax1.plot(J_history[:100])
ax2.plot(1000 + np.arange(len(J_history[1000:])), J_history[1000:])
ax1.set_title("Cost vs. iteration(start)") 
ax2.set_title("Cost vs. iteration (end)")
ax1.set_ylabel('Cost')
ax2.set_ylabel('Cost')
ax1.set_xlabel('iteration step')
ax2.set_xlabel('iteration step')
plt.show()
```

## Step 6: Plot to see the linear fit
* calculate all the predictions for final value of w and b

```python
m = len(x_train)

predicted = np.zeros(m)

for i in range (m):
    predicted[i] = w_final * x_train[i] + b_final 
```

* plot the linear fit 

```python
# Plot the linear fit
plt.plot(x_train, predicted, c = "b")

# Create a scatter plot of the data. 
plt.scatter(x_train, y_train, marker='x', c='r') 

# Set the title
plt.title("Profits vs. Population per city")
# Set the y-axis label
plt.ylabel('Profit in $10,000')
# Set the x-axis label
plt.xlabel('Population of City in 10,000s')
```

## Step 7: Make Predictions
*  Let's predict what the profit would be in areas of 35,000 and 70,000 people.

```python
predict1 = w_final * 3.5 + b_final
predict2 = w_final * 7 + b_final

print(f"For population {3.5 * 10000}, models prediction is : {predict1 * 10000}$")
print(f"For population {7 * 10000}, models prediction is   : {predict2 * 10000}$")
```
